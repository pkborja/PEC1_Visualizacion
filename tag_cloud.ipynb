```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
path = 'C:/Users/borja/Desktop/Academia/master/visualizacion/PEC2/words_en.csv'

df = pd.read_csv(path)
```

```python
data = df.loc[(df['word_type']!="IN") & (df['word_type']!="TO") & (df['word_type']!="PRP") &
               (df['word_type']!="PRP$") & (df['word_type']!="PRP$") & (df['word_type']!="DT") &
              (df['word_type']!="VBZ") & (df['word_type']!="VBD") & (df['word_type']!="EX") &
              (df['word_type']!="CC") & (df['word_type']!="VBP") & (df['word_type']!="RB")]
```

```python
records = data.to_dict(orient='records')

bible = {x['word']:x['number_of_uses'] for x in records}

wc = WordCloud(background_color = "white", max_words = 1000, width = 1000, height = 500)
wc = wc.generate_from_frequencies(bible)

plt.figure( figsize=(20,10), facecolor='k' )
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.savefig('tag_cloud.png')
plt.show()
```
